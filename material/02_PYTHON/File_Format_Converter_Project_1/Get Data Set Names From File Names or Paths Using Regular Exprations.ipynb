- Here we need to get all the Data set Names form the path and based on that Data set name we will read the data from the CSV file.

- In order to do that we can't use **split** function directly on the path because our path is having both **\ and //** so that we need to use **split function which takes regular exprations** as argumnent 

 - for that we need to use **re** module for regular exprations.

```python
import re
import glob
import pandas as pd 
import json         
```

```python
src_file_names = glob.glob(
    'E:/Projects/Data_Engineering/Data-Engineering/material/02_PYTHON/File_Format_Converter_Project_1/data/retail_db/*/part-*',
    recursive= True
)
src_file_names
```

```python
# split the paths to get all the Data set Name.
for file_path in src_file_names:
    data_set_names = re.split('[/\\\]', file_path)
    print(data_set_names)
```

## Read CSV Data into Pandas Dataframes with schema dynamically.

```python
file = src_file_names[0]
file
```

```python
data_set_name = re.split('[/\\\]',file)[-2]
data_set_name
```

```python
# Get all the column names for data set name 
file_path = 'E:/Projects/Data_Engineering/Data-Engineering/material/02_PYTHON/File_Format_Converter_Project_1/data/retail_db/schemas.json'
schema = json.load(open(file_path))

# get column details.

def get_column_name(schema, tableName, sortingKey='column_position'):
    column_details = schema[tableName]
    column_details_sort = sorted(column_details, key=lambda col:col[sortingKey])
    return [col['column_name'] for col in column_details_sort]


orders_column_names = get_column_name(schema, data_set_name)
orders_column_names
```

```python
data_set_name_data = pd.read_csv(
                                file,
                                names= orders_column_names
)
```

```python
data_set_name_data
```

### Combine all the details above for all the Data Sets

```python
for file in src_file_names:
    print(f'processing {file}')
    data_set_name = re.split('[/\\\]', file)[-2]
    data_set_columns = get_column_name(schema, data_set_name)
    df = pd.read_csv(
                    file,
                    names= data_set_columns
                    )
    
    print(f'shape of {data_set_name}\'s is {df.shape}')
    
```
