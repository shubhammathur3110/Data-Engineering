- First we will read the data and then apply the filters using query function

```python
import pandas as pd
column_list = ["order_id","order_date", "order_customer_id","order_status"]
file_path = 'E:/Projects/Data_Engineering/Data-Engineering/data/retail_db/orders/part-00000'
orders = pd.read_csv(
                    file_path,
                    names= column_list
)
```

- We need to filter the data based on the Order status = COMPLETE 
- First we will check the unique data for the order status.
- We will mostly use **boolen and or boolen or** for the filter.


```python
orders['order_status'].unique()
```

```python
orders.query('order_status == "COMPLETE"')
```

```python
orders.query('order_status == "COMPLETE" and order_date == "2014-01-01 00:00:00.0" ')
```

```python
orders.query('order_status == "COMPLETE" or order_status == "CLOSED"')
```

```python
orders.query('order_status == ("COMPLETE", "CLOSED")')
```
