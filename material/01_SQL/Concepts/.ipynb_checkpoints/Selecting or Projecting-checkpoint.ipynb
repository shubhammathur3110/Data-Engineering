## Project data with subset of the fields

```python
%%sql 

select order_customer_id, order_date, order_status
    from shubham.orders
        limit 10 ;
```

- **use to_char() function**

```python
%%sql
    select order_customer_id, 
        to_char(order_date, 'yyyy-MM'),
        order_status
    from shubham.orders
        limit 10;
```

## SQL Alias **'AS'**

   - We will define alias name after **AS** keyword
   - in the above example we can see that the date column name is **to_char**
   - Below you will see the name AS we define

```python
%%sql
    select order_customer_id, 
        to_char(order_date, 'yyyy-MM') AS order_month,
        order_status
    from shubham.orders
        limit 10;
```

## DISTINCT 

 - To get all the distinct values for the columns

```python
%%sql 

    select DISTINCT to_char(order_date, 'yyyy-MM') as order_month
    from shubham.orders;
```

## COUNT

- To see the count of Rows (perticular column or complete Table) 
- for count we will use **'COUNT()'**  function

### count for complete table

```python
%%sql
    select count(*) 
    from shubham.orders;
```

### count for perticular column 
   - with distinct values

```python
%%sql 
    select count(DISTINCT to_char(order_date, 'yyyy-MM')) AS distinct_month_count
    from shubham.orders;
```

- without distinct (all rows)

```python
%%sql 
    select count(to_char(order_date, 'yyyy-MM')) AS distinct_month_count
    from shubham.orders;
```
