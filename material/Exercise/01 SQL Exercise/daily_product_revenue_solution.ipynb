```python
  %%sql 

    select o.order_date,
           oi.order_item_product_id,
           p.product_name,
            round(sum (oi.order_item_subtotal:: numeric), 2) as Daily_Revenue
    from shubham.orders as o
        Inner join shubham.order_items as oi
            on o.order_id = oi.order_item_order_id
        Inner join shubham.products as p
            on  p.product_id = oi.order_item_product_id
    where o.order_status in ('COMPLETE', 'CLOSED')
    group by o.order_date, oi.order_item_product_id, p.product_name
    order by o.order_date, daily_revenue DESC
        limit 10; 
```
